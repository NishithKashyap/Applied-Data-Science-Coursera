```python
df2017 = df_final[df_final['season'] == 2017]
df2016 = df_final[df_final['season'] == 2016]
df2015 = df_final[df_final['season'] == 2015]
df2014 = df_final[df_final['season'] == 2014]
df2013 = df_final[df_final['season'] == 2013]
df2012 = df_final[df_final['season'] == 2012]
df2011 = df_final[df_final['season'] == 2011]
df2010 = df_final[df_final['season'] == 2010]
df2009 = df_final[df_final['season'] == 2009]
df2008 = df_final[df_final['season'] == 2008]
```

```python
def count_percentage(df_temp):
    count = df_temp['winner'].value_counts()
    return count
```

```python
count2017 = count_percentage(df2017)
count2016 = count_percentage(df2016)
count2015 = count_percentage(df2015)
count2014 = count_percentage(df2014)
count2013 = count_percentage(df2013)
count2012 = count_percentage(df2012)
count2011 = count_percentage(df2011)
count2010 = count_percentage(df2010)
count2009 = count_percentage(df2009)
count2008 = count_percentage(df2008)
```

```python
def rename_index(df_temp):
    df_temp = df_temp.rename(index = {'Kings XI Punjab':'KXIP','Mumbai Indians':'MI',
                                      'Delhi Daredevils':'DD','Kolkata Knight Riders':'KKR','Rajasthan Royals':'RR',
                                      'Royal Challengers Bangalore':'RCB','Deccan Chargers':'SRH','Deccan Chargers':'SRH',
                                      'Chennai Super Kings':'CSK','Pune Warriors':'PWI','Sunrisers Hyderabad':'SRH',
                                      'Rising Pune Supergiants':'PWI','Gujarat Lions':'GL','Rising Pune Supergiant':'PWI'})
    return df_temp
```

```python
year = [2017,2016,2015,2014,2013,2012,2011,2010,2009,2008]
num = [1,0,2,4,6,8,10,12]
colors = ['yellow','darkslateblue','lightsalmon','black','pink','blue','cyan','red','purple','orange']

ax = df_t.plot(color = colors)
ax.set_xticklabels(year,fontsize=20)
ax.set_yticklabels(num,fontsize=20)
plt.xlabel('Year',fontsize=28)
plt.ylabel('Number of wins', fontsize=28)
plt.legend(frameon=False, bbox_to_anchor=(1.05, 1),fontsize=20)
plt.figure (figsize=(10,5))
```
